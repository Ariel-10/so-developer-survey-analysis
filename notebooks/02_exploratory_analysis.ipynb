# Exploratory Analysis - Stack Overflow Developer Survey (2023-2025)

## Objective
- Explore the cleaned dataset before answering the main question (check the README)
- Look for outliers, missing data and basic patterns by role and year

```python
import pandas as pd

# Load cleaned data - Already unified across 3 years
df = pd.read_csv("../data/processed/survey_clean_2023_2025.csv")
df.head()
```

## 1.- Data Quality Check

```python
# Check which columns have mixed types
for col in df.columns:
    types = df[col].dropna().apply(type).unique()
    if len(types) > 1:
        print(col, types)
```

### YearsCode - Mixed Formats

Found while exploring: 
- Numbers as text
- Numbers as decimals
- Two special values ("Less than 1 year" and "More than 50 years").

Needs cleaning before we can chart this column.

```python
# Check unique YearsCode values with type
print(df["YearsCode"].unique())
```

```python
# Clean YearsCode - Unify text and mixed number formats
df["YearsCode"] = df["YearsCode"].replace({
    "Less than 1 year": 0,
    "More than 50 years": 51
})
df["YearsCode"] = pd.to_numeric(df["YearsCode"], errors="coerce")

print(df["YearsCode"].describe())
```

```python
# Check how many responses claim 40+ years coding
print((df["YearsCode"] > 40).sum())
```

```python
# Check how many responses reach more than 60 years (probably bad data, not just long careers)
print((df["YearsCode"] > 60).sum())
```

#### Findings - YearsCode
- Found 71 responses reaching more than 60 years coding - Probably bad data (not real careers)

We should **remove** it before chart.

```python
# YearsCode > 60 is not physically possible given the survey's max age bracket
# Set to NaN instead of dropping the row - Keeps the rest of that person's data
df.loc[df["YearsCode"] > 60, "YearsCode"] = pd.NA

print(df["YearsCode"].describe())
```

### ConvertedCompYearly - Check for Outliers
Salary data in surveys is usually messy because people put the number in the wrong currency or write extra zeros by mistake.


```python
print(df["ConvertedCompYearly"].describe())
```

```python
# Check extreme values on both ends
print((df["ConvertedCompYearly"] > 1_000_000).sum())
print((df["ConvertedCompYearly"] < 1_000).sum())
```

```python
# Check countries with high salary responses
print(df[df["ConvertedCompYearly"] > 1_000_000]["Country"].value_counts().head(10))
```

```python
# Limit salary outliers - Values above one million could be currency or typo errors (not real salaries)
df.loc[df["ConvertedCompYearly"] > 1_000_000, "ConvertedCompYearly"] = pd.NA

print(df["ConvertedCompYearly"].describe())
```

#### Findings - High salary check

**Found:** 147 responses over one million dollars per year

**Checked:** by country

**Pattern:** No economic reason for these amounts - They are spread across countries where one million dollars salaries are not realistic.

**Conclusion:** Probably currency or typo errors - Set NaN

```python
# Check countries with low salary responses
print(df[df["ConvertedCompYearly"] < 1000]["Country"].value_counts().head(10))
```

```python
# Check how low the low salaries are
print(df[df["ConvertedCompYearly"] < 1000]["ConvertedCompYearly"].describe())
```

```python
# Check how many fall below a realistic minimum (100 dollars per year)
print((df["ConvertedCompYearly"] < 100).sum())
```

```python
# Below 100 dollars per year is not a realistic salary anywhere
df.loc[df["ConvertedCompYearly"] < 100, "ConvertedCompYearly"] = pd.NA

print(df["ConvertedCompYearly"].describe())
```

#### Findings - Low salary check:

**Found:** 1953 responses under $1000 per year

**Checked:** by country - Low cost of living countries
(Ukraine, India, Nigeria, Pakistan)

**Looked closer:** 720 of those are under $100 per year - There is not any country with real salary as low as that

**Conclusion:** 
- Below 100 dollars set NaN (bad data)
- Between 100 - 1000 dollars kept as real salaries.

## 2.- Missing Data Check

**How complete is each column?** - It helps to decide what is usable for charts later.

```python
# Check missing AI data - Only in the 4 roles that we are interested for the main question
role_subset = df[df["RoleClean"].notna()]
print(role_subset.shape)
print(role_subset[["AISelect", "AISent", "AIAcc"]].isnull().sum())
```

## 3.- Main Question - AI Adoption by Role

**First look:** AISelect, AISent and AIAcc by role across the 3 years.

### AISelect - Uses AI?

```python
# AISelect by role and year - Uses AI?
print(role_subset.groupby(["RoleClean", "Year"])["AISelect"].value_counts(normalize=True)) # Percentages
```

#### Findings - AISelect

**Found:** Data Analyst behind Data Engineer and Data Scientist - daily use

**2025 daily use:** Data Analyst 36% - Data Engineer 52% - Data Scientist 52%

**Database Administrator:** last place (24% daily)

### AISent - How favorable toward AI?

```python
# AISent by role and year - How favorable toward AI?
print(role_subset.groupby(["RoleClean", "Year"])["AISent"].value_counts(normalize=True))
```

```python
# % favorable or very favorable toward AI by role and year
favorable = role_subset[role_subset["AISent"].isin(["Favorable", "Very favorable"])]
print(favorable.groupby(["RoleClean", "Year"]).size() / role_subset.groupby(["RoleClean", "Year"])["AISent"].count())
```

#### Findings - AISent
**Found:** Favorable/very favorable drops from 2023  to 2025 - All 4 roles

**Data Analyst:** From 80% to 63%

**Pattern:** Same drop across all roles - Not just one

### AIAcc - Trust in AI Accuracy?

```python
# % Who trust AI accuracy by role and year - Highly trust or Somewhat trust
trust = role_subset[role_subset["AIAcc"].isin(["Highly trust", "Somewhat trust"])]
print(trust.groupby(["RoleClean", "Year"]).size() / role_subset.groupby(["RoleClean", "Year"])["AIAcc"].count())
```

#### Findings - AIAcc

**Found:** Trust in AI accuracy - Drop pattern as AISent

**Data Analyst:** 2024 (55%) - 2025 (44%)

**Database Administrator:** Biggest drop - From 61% to 31%

### AIComplex - Trust for Complex Tasks?

```python
# Check AIComplex categories first
print(role_subset["AIComplex"].unique())
```

```python
# % Who rate AI as good or very well for complex tasks by role and year
complex_trust = role_subset[role_subset["AIComplex"].isin([
    "Very well at handling complex tasks",
    "Good, but not great at handling complex tasks"
])]
print(complex_trust.groupby(["RoleClean", "Year"]).size() / role_subset.groupby(["RoleClean", "Year"])["AIComplex"].count())
```

#### Findings - AIComplex

**Found:** Trust drops from 2024  to 2025 - All 4 roles

**Data Analyst:** From 47% to 35%

**Risk:** DBA drop from 57% to 24% - Small sample (n=70 in 2024)

### AIAgents - Using AI Agents?

```python
# Check AIAgents categories first
print(role_subset["AIAgents"].unique())
```

```python
# % Using AI agents (daily, weekly or monthly) by role - 2025 only
using_agents = role_subset[role_subset["AIAgents"].isin([
    "Yes, I use AI agents at work daily",
    "Yes, I use AI agents at work weekly",
    "Yes, I use AI agents at work monthly or infrequently"
])]
print(using_agents.groupby("RoleClean").size() / role_subset.groupby("RoleClean")["AIAgents"].count())
```

#### Findings - AIAgents

**Found:** Data Analyst 31%  -  Data Scientist 31%

**Pattern:** Different from AIComplex and AISent - Data analyst followed both of them.

**Note:** Data Engineer close (30%) - Database Administrator last (22%)
